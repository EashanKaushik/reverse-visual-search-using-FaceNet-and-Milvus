# FaceNet - Model Initialize

```python
import tensorflow as tf
from tensorflow.keras import models
from keras.models import Model

model_path = "FaceNet/model/facenet_keras.h5"
weight_path = "FaceNet/weights/facenet_keras_weights.h5"

model = models.load_model(model_path, compile=False)

model.load_weights(weight_path)
```

```python
model.summary()
```

### Adding a Avg Pooling Layer

```python
layer_name = 'AvgPool'
facenet= Model(inputs=model.input, outputs=model.get_layer(layer_name).output)
```

# Read Dataset

```python
from matplotlib import image
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
X = np.load(os.path.join("query_images", 'query_images.npy'))
```

```python
X.shape
```

```python
from PIL import Image

X_reshape = list()
count = 1
for index in range(X.shape[0]):
  image = Image.fromarray(X[index])
  image = image.resize((160, 160))
  face_array = np.asarray(image)
  X_reshape.append(face_array)
  print(f'Resized {count}')
  count += 1

X_reshape = np.array(X_reshape)
```

```python
X_reshape.shape
```

# Preprocess

```python
def preprocess(face):
    # scale pixel values
    face = face.astype('float32')
    # standardization
    mean, std = face.mean(), face.std()
    face = (face-mean)/std
    return face
```

```python
X_preprocess = list()
count = 1
for train in X_reshape:
  X_preprocess.append(preprocess(train))
  print(f'pre process {count}')
  count += 1
X_preprocess = np.array(X_preprocess)
```

```python
X_preprocess.shape
```

# Features

```python
X_features = facenet.predict(X_preprocess)
```
