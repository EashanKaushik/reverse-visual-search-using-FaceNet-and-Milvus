# FaceNet

```python
import tensorflow as tf
from tensorflow.keras import models
from keras.models import Model

model_path = "0004/FaceNet/model/facenet_keras.h5"
weight_path = "0004/FaceNet/weights/facenet_keras_weights.h5"

model = models.load_model(model_path, compile=False)

model.load_weights(weight_path)
```

```python
model.summary()
```

```python
layer_name = 'AvgPool'
facenet= Model(inputs=model.input, outputs=model.get_layer(layer_name).output)
```

```python
from matplotlib import image
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
train = pd.read_csv(os.path.join("0001", "train.csv"), index_col=0)
test = pd.read_csv(os.path.join("0001", "test.csv"),  index_col=0)
labels = pd.read_csv(os.path.join("0001", "labels.csv"), index_col=0)
```

```python
X_train = list()
for name in train["Name"]:
    directory = name.split("0")[0][: -1]
    X_train.append(image.imread(os.path.join("LFW", directory, name)))
X_train = np.array(X_train)
```

```python
X_test = list()
for name in test["Name"]:
    directory = name.split("0")[0][: -1]
    X_test.append(image.imread(os.path.join("LFW", directory, name)))
X_test = np.array(X_test)
```

```python
y_train = np.array(train["Label"])
y_test = np.array(test["Label"])
```

```python
X_train_reshape = list()

for index in range(X_train.shape[0]):
    X_train_reshape.append(tf.image.resize(X_train[index], [160, 160]))

X_train_reshape = np.array(X_train_reshape)
```

```python
X_test_reshape = list()

for index in range(X_test.shape[0]):
    X_test_reshape.append(tf.image.resize(X_test[index], [160, 160]))

X_test_reshape = np.array(X_test_reshape)
```

```python
X_train_features = facenet.predict(X_train_reshape)
```
